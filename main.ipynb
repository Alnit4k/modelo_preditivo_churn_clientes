Para o nosso projeto usaremos uma base de churn,  essa base se diz respeito a serviços de internet, telefone e tv - Telecomunicação.

Legenda dos dados:

*   **CustomerID** : Esse é o ID único de cada cliente nosso da rede de produtos de telecomunicação.

*   **Servico_Internet** : Qual o tipo de serviço de internet contratado.

*   **Servico_Seguranca** : Se ele contratou serviço de segurança com a nossa empresa.

*   **Suporte_Tecnico** : Se já acionou o suporte técnico.

*   **StreamingTV** : Se ele contratou serviço de streaming para TV.

*   **Tipo_Contrato** : Qual o tipo de contrato dos nossos clientes, se era mensal, anual etc.

*   **Tipo_Contrato** : Qual o tipo de contrato dos nossos clientes, se era mensal, anual etc.

*   **PaymentMethod** : Forma de Pagamento.

*   **Pagamento_Mensal** : Quanto o cliente pagava mensalmente.

*   **Total_Pago** : Total já pago pelo cliente para nossa companhia.

*   **Churn** : Nossa coluna mais importante, se nosso cliente nos abandonou ou não.

### Tratamento de dados nulos


Optei por deletar a coluna [PhoneService],  pois aferindo que nao possui relevancia significativa para a nossa analise, alem disso se trata de uma coluna boleana, por padrao os dados nulos seriam preenchidos com false, mas isso tornaria minha analise enviesada e com possiveis outliers pelo alta quantidade de dados nulos.



```python
df['PhoneService'].unique()
```

```python
df.drop('PhoneService', axis=1)
```

Indo para a coluna [Pagamento mensal] podemos observar que se trata de uma coluna com poucos dados faltantes, representando 13% de todo conjunto, e a media nao e discrepante em relacao aos valores maximos e minimos como pudemos observar pela funcao describe. Por se tratar de uma coluna importante para analise, pois o pagamento pode influenciar na presenca ou saida do cliente optei por preencher os valores nulos pela media da coluna.


```python
df['Pagamento_Mensal'].describe()
df['Pagamento_Mensal'] = df['Pagamento_Mensal'].fillna(df['Pagamento_Mensal'].mean())
```

Sobre a coluna [Genero] ela tem um numero insigficante de valores nulos representando apenas 0.48% isso equivalore a apenas 12 linhas. Logo aferi que a quantidade de valores nulos e extremamente insignificante na coluna genero, alem disso nao se trata de uma analise que o genero impacta substancialmente, por isso optei em apagar as linhas que contem os valores nulos.

```python
df = df.dropna(subset=['Genero'])
df['Genero'].isnull().sum()
```

```python
# Padronizando os valores categoricos
df['Genero'].unique()
mapa = {'Female' : 'F', 'Male': 'M', 'f': 'F'}
df['Genero'] = df['Genero'].map(mapa)
```

```python
# Padronizando colunas
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)
```
